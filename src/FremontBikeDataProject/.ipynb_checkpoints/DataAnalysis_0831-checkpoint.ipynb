```python
from urllib.request import urlretrieve
import os
import pandas as pd

URL = "https://data.seattle.gov/api/views/65db-xm6k/rows.csv?Type=DOWNLOAD"

def get_fremont_data(filename = 'Fremont.csv', url=URL, force_download=False ):

    """Download and chach the Fremont data

    Parameters
    --------------
    filename : string (optional)
        location to see the data
    url : sting (optional)
        web location to download the file
    force_download: bool (optional)
        if True, force download of the data

    Return
    -------
    csv_data : panda dataframe

    """

    if force_download or not os.path.exists("Fremont.cav"):
        urlretrieve(URL, 'Fremont.csv')
    
    csv_data = pd.read_csv('Fremont.csv', index_col = 'Date', parse_dates = True)
    return csv_data

test = get_fremont_data()
```
